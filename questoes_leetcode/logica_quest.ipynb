### Add Two Numbers
---

<span style="color:#2563eb;">Nível: Médio</span>

#### Descrição
Você recebe **duas listas encadeadas não vazias** que representam dois números inteiros não negativos.

- Os dígitos estão armazenados em **ordem reversa**.
- Cada nó contém **um único dígito**.
- Some os dois números e retorne o resultado como **lista encadeada**.

### <span style="color:#f59e0b;">Observação</span>
Pode assumir que os dois números **não possuem zeros à esquerda**, exceto o número `0`.



### <span style="color:#7c3aed;">Exemplo</span>

**Input:** `l1 = [2,4,3]`, `l2 = [5,6,4]`  
**Output:** `[7,0,8]`  
**Explicação:** `342 + 465 = 807`

> inicio da minha lógica.

#### Como percorrer uma linked list?



> pensar: você vai de elemento em elemento

Então, não seria p1[i] e sim p1.val

não é p1[i+1], é p1=p1.next

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        carry = 0
        p1,p2 = l1,l2
        dummy = ListNode(0)
        current=dummy

        while p1 or p2 or carry:
            val1=p1.val if p1 else 0
            val2=p2.val if p2 else 0

            soma=val1+val2+carry
            carry = soma//10
            digit=soma%10
            current.next = ListNode(digit)
            current = current.next

            if p1:
                p1=p1.next
            if p2:
                p2 = p2.next
        
        
        return dummy.next
```

### Merge Two Sorted Lists (LeetCode)
---
Easy

You are given the heads of two sorted linked lists list1 and list2. Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.
Return the head of the merged linked list.

```python
class Solution:
    def mergeTwoLists(self, list1, list2):
        dummy = ListNode(0)
        atual = dummy
        p1=list1
        p2=list2

        while p1 and p2:
            if p1.val<=p2.val:
                atual.next = p1
                atual=atual.next
                p1 = p1.next
            else:
                atual.next = p2
                atual = atual.next
                p2=p2.next
        if p1:
            atual.next = p1

        else:
            atual.next=p2

        return dummy.next 
```

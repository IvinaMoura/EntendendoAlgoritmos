> **Importante ressaltar**: a recurso  usada para tornar a reposta mais clara. No h nenhum benefcio quanto ao desempenho ao utilizar recurso.

```python
def regressiva(i):
    print(i)
    return regressiva(i-1) # essa funo executar para sempre!
```

Toda funo recursiva tem duas partes: o caso-base e o caso recursivo.
- Caso recursivo: quando a funo chama a si mesma.
- Caso-base: quando a funo no chama a si mesma novamente, de forma que o programa no se torna um loop infinito.

```python
# add caso base  funo de contagem regressiva
def regressiva(i):
    print(i)
    if i<=1: # caso-base
        return
    else: 
        regressiva(i-1) # caso recursivo
```

- Agora a funo vai funcionar como o esperado.

### Arrays e listas encadeadas


1. Usar um array significa que todas as suas tarefas esto armazenadas contiguamente (uma ao lado da outra) na memria.





* Listas encadeadas: seus itens podem estar em qualquer lugar da memria.


Cada item armazena o endereo do prximo item da lista. Um monte de endereos aleatrios de memria esto ligados.



> Listas encadeadas so timas se voc quiser ler todos os itens, um de cada vez: ler um item, v o endereo do prximo e assim sucessivamente. MAS se quiser pular de um item para outro, as listas encadeadas so pssimas.



> Arrays so timos se voc deseja ler elementos aleatrios, pois pode encontrar qualquer elemento instantaneamente em um array.

Tempo de execuo para operaes comuns de arrays e listas:



- Listas encadeadas so melhores caso queira inserir um elemento no meio de uma lista. 
    - basta mudar o endereo para o qual o elemento anterior est apontando. 
    - J nos arrays voc deve mover todos os itens que esto abaixo do endereo de insero.

- Delees:
    -  mais fcil fazer com listas encadeadas, mudar apenas o endereo para o qual o elemento anterior est apontando.
    - Arrays, tudo precisa ser movido quando um elemento  eliminado.

Tempo de execuo para as operaes mais comuns em arrays e listas encadeadas.



```python
# Funo para encontrar o menor elemento em um array:

def buscaMenor(arr):
    menor = arr[0] # armazena o menor valor
    menor_indice = 0 # armazena o ndice do menor valor

    for i in range (1,len(arr)):
        if arr[i] < menor:
            menor=arr[i]
            menor_indice=i
    return menor_indice
```

```python
# ordenao por seleo
def ordenacaoporSelecao(arr):
    novoArr=[]
    for i in range(len(arr)):
        menor = buscaMenor(arr)
        novoArr.append(arr.pop(menor))
    return novoArr

print(ordenacaoporSelecao([5,3,6,2,10]))
```

Trabalha em um mercado, precisa conferir o preo de um produto em um caderno. Porm, se o caderno no estiver organizado alfabeticamente, far uma busca simples (vista no cap. 1 O(n)).

No entanto, se o caderno estiver ordenado alfabeticamente poderia executar uma pesquisa binria



No entanto, mais interessante que busca binria seria um colega de trabalho que soubesse o preo de todos os produtos, voc pergunta -> ela informa o preo imediatamente.



A sua amiga Maggie pode dizer o preo com tempo de execuo O(1) para todos os itens, no importando a quantidade de itens que compem o caderno de preos.

#### Funes hash
---
Uma funo hash  uma funo na qual voc insere uma string e, depois disso, a funo retorna um nmero.




> Em uma terminologia mais tcnica: uma funo hash "mapeia strings e nmeros".

Existem alguns requisitos para uma funo hash:
- Ela deve ser consistente. Se inserir uma string "ma" e receber o nmero 4. Todas as vezes que voc inserir "ma", a funo deve retornar o nmero 4. Caso contrrio, sua tabela hash no funcionar corretamente.

#### Funções hash
---
Uma função hash é uma função na qual você insere uma string e, depois disso, a função retorna um número.




> Em uma terminologia mais técnica: uma função hash "mapeia strings e números".

Existem alguns requisitos para uma função hash:
- Ela deve ser consistente. Se inserir uma string "maçã" e receber o número 4. Todas as vezes que você inserir "maçã", a função deve retornar o número 4. Caso contrário, sua tabela hash não funcionará corretamente.

Tabelas hash também são conhecidas como: mapas hash, mapas, dicionários e tabelas de dispersão. 

> Provavelmente nunca teremos que implementar uma tabela hash, pois qualquer linguagem de programação já terá uma implementação dela. A linguagem Python contém tabelas hash chamadas de dicionários.

```python
caderno = dict()
# caderno é uma nova tabela hash, agora, vamos adicionar alguns preços nela.
caderno['maca'] = 0.67
caderno['banana'] = 0.34
caderno['leite']=1.49
caderno['abacate'] = 1.49
print(caderno)
```

```python
print(caderno['abacate'])
```

Na hash caderno, os nomes dos produtos são as chaves e os preços são os valores. Logo, uma tabela hash mapeia chaves e valores.


```python
# exemplo de utilização:
lista_telefonica = dict()
# ou
lista_telefonica = {}
lista_telefonica['maria'] = '555-1234'
lista_telefonica['joao'] = '555-5678'
lista_telefonica['jenny'] = '0867-5309'
print(lista_telefonica['maria'])
```

#### Exemplo de uso de dicionário para evitar entradas duplicadas (votação)
---


```python
votaram={}
valor = votaram.get('tom')
print(valor)
```

```python
voted = {}

def verifica_eleitor(nome):
    if votaram.get(nome):
        print(f'Eleitor {nome} já votou.')
    else:
        voted[nome]=True
        print(f'Eleitor {nome} votou com sucesso.')

# testes
verifica_eleitor('mike')
print(voted)
```

#### Utilizando tabelas hash como cache
---
Último uso para tabelas hash: utilização como cache. 



Como cache funciona: Os websites lembram dos dados em vez de recalculá-los a cada solicitação.



Isso é *caching*, e esta prática oferece duas vantagens:
- Você recebe a página web mais rapidamente, da mesma forma que você memorizou a distância entre a Terra e a Lua.
- O facebook precisa trabalhar menos.

> Todos os grandes sites usam caching, e os dados destes cachings são armazenados em uma tabela hash!

### Recapitulando
As tabelas hash são úteis para
- Modelar relações entre dois itens
- Filtrar por dupllicatas
- Caching/memorização de dados, em vez de solicitar estes dados do servidor.


### Colisões
---
Na realidade, é praticamente impossível escrever uma função hash que faça mapeamento diferente em chaves para diferentes espaços em um array.

- Duas chaves são indicadas para o mesmo espaço, isto é um problema (colisão).
    - No exemplo, tanto "ameixa" quanto o "abacate" são mapeados para o mesmo espaço.
    - Para solucionar o problema, a mais simples é: se diversas chaves mapeiam para o mesmo espaço, inicie uma lista encadeada neste espaço.



### Desempenho
---

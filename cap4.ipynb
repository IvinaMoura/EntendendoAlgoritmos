### Dividir para conquistar


### Exemplo 1:
---
Fazendeiro que  quer dividir sua fazenda em porções quadradas iguais, sendo que estas porções devem ter o maior tamanho possível.


Para resolver um problema utilizando DC(dividir para conquistar), você deve seguir os dois passos:
1. Descubra o caso-base, que deve ser o caso mais simples possível.
2. Divida ou diminua o seu problema até que ele se torne o caso-base.

Seria mais fácil se um ds lados fosse múltiplo do outro:

O maior quadrado que você pode ter mede 25m x 25m. Necessário dois passos para dividir a porção de terra.

Agora é necessário descobrir o caso recursivo.




#### Algoritmo de Euclides
>"Caso você encontre o maior quadrado que divide este segmento, ele será o maior quadrado que irá dividir toda a fazenda."

### Exemplo 2
---
Você tem um array de números [(2,4,6)] -> você deve somar todos os números e retornar o valor total.

```python
def soma(lista):
    total =0
    for x in lista:
        total += x
    return total

print(soma([1,2,3,4]))
```

> Mas como poderia ser feito com uma função recursiva?

**Passo 1:** Descubra o caso-base. Qual array mais simples que é possível obter?


**Passo 2:** Você deve chegar mais perto ed um array vazio a cada recursão.


Na segunda versão está sendo utilizado o array menor na função soma >>> diminuindo o tamamnho do problema.



### Quicksort
---

```python
def quicksort(array):
    if len(array) < 2:
        return array

    # um array com dois elementos é simples de ordenar
    elif array[0] < array[1]:
        return array
    
```

O funcionamento do quicksort segue essa lógica: primeiro, escolha um elemento do array. Esse elemento será chamado de pivô.






Isso é chamado de particionamento.
- subarray contendo todos os números menores do que o pivô.
- O pivô
- Um subarray contendo todos os números maiores do que o pivô.

Se os subarrays estiverem ordenados: array esquerdo + pivô + array direito.

O caso-base do quicksort consegue ordenar arrays de dois elementos (o subarray esquerdo) e também arrays vazios (subarray direito).
> quicksort([15,10]) + [33] + quicksort([]) -> [10,15,33]

- Passos:
---
1. Escolha um pivô
2. particione o array em dois subarrays, separando-os entre elementos menores do que o pivô e elementos maiores do que o pivô.
3. execute quicksort recursivamente em ambos os subarrays.
